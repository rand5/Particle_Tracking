In each of the above plots, we observe one particle that appears to exhibit behavior distinct from that of the other particles. To determine if this difference is statistically distinct, we will apply a two-sample Kolmogorov-Smirnov (KS) test. The KS test is a non-parametric test of the null hypothesis that two samples are drawn from the same distribution against the alternative hypothesis that the distribution comes from different distributions.

In preparation for this test, we will select three particle paths, one of which is the particle we suspect to be in the organoid ($ID=1$), and compute their displacements $d_i$, where $d_i = x_{i}-x_{i-1}$

```julia
Path1 = April1_DF[end,:Path]; # Inspecting the output of PBA_Particle_Loader shows that the particle with ID = 1 was loaded last.
Path1 = diff(Path1.x);

Path2 = April1_DF[2,:Path];
Path2 = diff(Path2.x);

Path3 = April1_DF[3,:Path];
Path3 = diff(Path3.x);
```

```julia
ApproximateTwoSampleKSTest(Path1,Path2)
```

```julia
ApproximateTwoSampleKSTest(Path1,Path3)
```

```julia
ApproximateTwoSampleKSTest(Path2,Path3)
```

According to the above results, the displacements exhibited by Particle 1, which we believed to be the particle within the organoid based on a visual inspection of the composite fluorescence and brightfield images, are highly likely to have come from a distribution distinct from that of Particle 2. A similar conclusion was reached when comparing the displacements of Particle 1 with those of Particle 3. We are unable to reject the null hypothesis when comparing the displacements of Particle 2 with those of Particle 3.
